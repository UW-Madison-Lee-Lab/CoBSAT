```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
def get_image(model, num_demos, misleading = False, other = '', endwith = ''):
    fix = '_m' if misleading else ''
    item = f'results/{model}_results/shot_{num_demos}{fix}'
    if other:
        item += f'/{other}'
    while os.path.isdir(item):
        # randomly pick one item from the folder
        item_list = os.listdir(item)
        
        if endwith:
            new_item_list = []
            for path in item_list:
                if path.endswith(endwith):
                    new_item_list.append(path)
            item_list = new_item_list
            
        selected_item = np.random.choice(item_list)
        
        while selected_item.startswith('.') or selected_item.startswith('_'):
            selected_item = np.random.choice(item_list)
        item = os.path.join(item, selected_item)
    
    # read the image
    image = plt.imread(item)
    return image
```

```python
# Assuming each image is a numpy array of shape (height, width, channels)
# Create 50 dummy images for demonstration
dummy_images = [None for _ in range(50)]
np.random.seed(0)

# Number of rows and columns for the big image
n_rows = 5
n_cols = 5
model, num_demos, misleading, other, endwith = 'gill', 1, 0, 'color_object/color_flower', 'green_result.jpg'

# Create a figure with subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 2, n_rows * 2))
fig.subplots_adjust(left=0, right=1, top=1, bottom=0, wspace=0, hspace = 0) #margin of the figure

for i, ax in enumerate(axes.flatten()):
    if i < len(dummy_images):
        ax.imshow(get_image(model, num_demos, misleading, other, endwith))
        ax.axis('off')  # Turn off axis
    else:
        ax.axis('off')  # Ensure no empty subplot has visible axis

plt.tight_layout(pad=0)
plt.show()
```
